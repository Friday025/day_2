```python
from tabulate import tabulate
import sys
import warnings
warnings.filterwarnings("ignore")




print("welcome to the restaurant")
# show menue and starter
starter_food = [['Soups',10], ['salads',15], ['bruschetta',20], ['potato skins',30], ['fried calamari',25] , ['shrimp',15]]

menue = {

    "Indian":{
        "Aloo Gobi": 80,
        "Butter Tikka" :60,
        "Chole Bhature" : 100,
        "Biryani":120,
        "Vada Pav": 50,
        "Chicken": 100,
        "Pasta": 70,
        "Pizza": 60,
        "Burger": 20,
        "Fries": 80
        },
    
    "chinese" :{
        "Chow Mein" :100,
        "Fried Rice": 70,
        "Fujian Red Wine Chicken":90,
        "Soy Egg" :60
    },

    "italian" :{
        "pizza":45,
        "pasta":58,
        "Franch Rice": 26,
        "Raspberry": 59
    }

}

table = []

starter = [["Starter", item, price] for item , price in starter_food]

starter_table = tabulate(starter, headers=['Category', 'Item', 'Price'], tablefmt="grid")


for food , item in menue.items():
    table.extend([[food, item,price] for item , price in item.items()])

table =  tabulate(table, headers= ["food", "items","price"], tablefmt="grid")



bill = 0
person = 0
# ask customer what they want to eat 
print("hello sir how can i help you .")
customer = input("i want to : ")



if customer == "starter":
    print(starter_table)

elif customer == "menue":
    print(f"here is your menue sir \n",table)


people = int(input("how many people are on the table : "))
person += people

if people > 6:
    print(f"sorry {people} people are not allowed our table limit is 5.")
    sys.exit()

is_order = True


while is_order:

    orders = input("What would you like to order? (Enter comma-separated items): ").title().split(',')
    
    for order in orders:
        found = False
        for cuisine in menue:
            if order.strip() in menue[cuisine]:
                print(f"Here is your {cuisine} dish: {order.strip()} - ${menue[cuisine][order.strip()]}")
                bill += menue[cuisine][order.strip()]
                found = True
                break
        if not found:
            print(f"Sorry, {order.strip()} is not available.")
        

    additional_order = input("would you like to add another order (yes/no)? ")
    if additional_order == "no":
        is_order = False
 
print(f"your bill is {bill*person}")
# ask tip for customer
weiter = int(input("what percentage of bill you want to give tip choose($5 to $20) : "))
print(f"Thank you for ${(bill*person)*weiter/100} tip here is your bill ${(bill*person)}")
print("Thank you sir visit again !")
is_order = False

```
